Find your country on a Map

```python
import plotly.express as px

country  = input("Enter your Country Name:")

data = {
    'country':[country],
    'Values':[100]
}
fig = px.choropleth(
    data,
    locations = 'country',
    locationmode='country names',
    color = 'Values',
    color_continuous_scale = 'Inferno',
    title = f'Country Map Highlighting {country}'
)

fig.show()
```

Chess Board Using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an 8x8 pattern (0 and 1)
chessboard = np.add.outer(range(8), range(8)) % 2

# Display with black & white squares
plt.imshow(chessboard, cmap='binary')
plt.xticks([]); plt.yticks([])  # Remove axis ticks
plt.title("Chess Board", fontsize=14)
plt.show()
```
